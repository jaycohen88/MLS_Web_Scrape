```python
import numpy as np
import pandas as pd
```

```python
MLS_data = pd.read_csv('MLS_data.csv', header = None)
```

```python
df1 = MLS_data.T
```

```python
data2019 = df1.iloc[:,:14]
```

```python
data2018 = df1.iloc[:-1,14:]
```

```python
data2019.iloc[:,0] = data2019.iloc[:,0] + ' 2019'
```

```python
data2018.iloc[:,0] = data2018.iloc[:,0] + ' 2018'
```

```python
data2019.columns=['Team','Pl','W','D','L','GS','GD','xGS','xGD','Cr','Comp_Cr','Press','Succ_Press','High_Press']
```

```python
data2018.columns=['Team','Pl','W','D','L','GS','GD','xGS','xGD','Cr','Comp_Cr','Press','Succ_Press','High_Press']
```

```python
TwoYearData = pd.concat([data2019, data2018], axis = 0)
```

```python
TwoYearData
```

```python
TwoYearData = TwoYearData.reset_index()
```

```python
TwoYearData = TwoYearData.drop('index', axis=1)
```

```python
TwoYearData[['Pl','W','D','L','GS','GD','xGS','xGD','Cr','Comp_Cr','Press','Succ_Press','High_Press']] = \
TwoYearData[['Pl','W','D','L','GS','GD','xGS','xGD','Cr','Comp_Cr','Press','Succ_Press','High_Press']].apply(pd.to_numeric)
```

```python
TwoYearData
```

```python
points = (TwoYearData['W'])*3 + (TwoYearData['D'])
```

```python
len(points)
```

```python
TwoYearData.insert(2, 'P', points)
```

```python
TwoYearData
```

```python
TwoYearData.to_csv('MLS_data_cleaned.csv', index=False)
```
